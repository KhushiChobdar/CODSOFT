```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Sample dataset (movies with genres)
movies = [
    {"title": "Avengers", "genre": "action superhero"},
    {"title": "Batman", "genre": "action superhero dark"},
    {"title": "Superman", "genre": "action superhero"},
    {"title": "Titanic", "genre": "romance drama"},
    {"title": "Notebook", "genre": "romance love"},
    {"title": "Inception", "genre": "sci-fi thriller"},
    {"title": "Interstellar", "genre": "sci-fi space"},
    {"title": "Joker", "genre": "drama dark"},
]

# Convert genres to vectors
def create_similarity_matrix(movies):
    genres = [movie["genre"] for movie in movies]
    cv = CountVectorizer()
    matrix = cv.fit_transform(genres)
    similarity = cosine_similarity(matrix)
    return similarity

# Recommend movies
def recommend(movie_name):
    similarity = create_similarity_matrix(movies)
    titles = [movie["title"] for movie in movies]

    if movie_name not in titles:
        print("Movie not found!")
        return

    index = titles.index(movie_name)
    scores = list(enumerate(similarity[index]))

    # Sort based on similarity score
    sorted_scores = sorted(scores, key=lambda x: x[1], reverse=True)

    print(f"\nRecommended movies for '{movie_name}':\n")

    # Show top 3 recommendations (excluding itself)
    count = 0
    for i in sorted_scores[1:]:
        print(titles[i[0]])
        count += 1
        if count == 3:
            break

# Run program
movie_input = input("Enter a movie name: ")
recommend(movie_input)
```
